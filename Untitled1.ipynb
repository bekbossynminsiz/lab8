```python
#task1 lab8
#imports
import pygame, sys
from pygame.locals import *
import random, time
import os

# initializing 
pygame.init()

# setting up FPS 
FPS = 60
FramePerSec = pygame.time.Clock()

# creating colors
BLUE  = (0, 0, 255)
RED   = (255, 0, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)

# screen settings
SCREEN_WIDTH = 400
SCREEN_HEIGHT = 600
SPEED = 5
SCORE = 0

# setting up fonts
font = pygame.font.SysFont("Verdana", 60)
font_small = pygame.font.SysFont("Verdana", 20)
game_over = font.render("Game Over", True, BLACK)

# checking if files exist before loading
if not os.path.exists("AnimatedStreet.png"):
    print("Error: Background image not found!")
    sys.exit()

if not os.path.exists("Enemy.png"):
    print("Error: Enemy image not found!")
    sys.exit()

if not os.path.exists("Player.png"):
    print("Error: Player image not found!")
    sys.exit()

if not os.path.exists("coin.png"):
    print("Error: Coin image not found!")
    sys.exit()

if not os.path.exists("crash.wav"):
    print("Warning: crash.wav not found!")

# loading images
background = pygame.image.load("AnimatedStreet.png")

# create a white screen 
DISPLAYSURF = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
DISPLAYSURF.fill(WHITE)
pygame.display.set_caption("Game")

# Enemy class
class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        self.image = pygame.image.load("Enemy.png")
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)  

    def move(self):
        global SCORE
        self.rect.move_ip(0, SPEED)
        if self.rect.top > SCREEN_HEIGHT:
            SCORE += 1
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

# Player class
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        self.image = pygame.image.load("Player.png")
        self.rect = self.image.get_rect()
        self.rect.center = (160, 520)
        
    def move(self):
        pressed_keys = pygame.key.get_pressed()
        if pressed_keys[K_LEFT] and self.rect.left > 0:
            self.rect.move_ip(-5, 0)
        if pressed_keys[K_RIGHT] and self.rect.right < SCREEN_WIDTH:
            self.rect.move_ip(5, 0)

# Coin class
class Coin(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("coin.png")
        self.image = pygame.transform.scale(self.image, (30, 30))
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), random.randint(40, SCREEN_HEIGHT - 40))

    def move(self):
        self.rect.move_ip(0, 2)
        if self.rect.top > SCREEN_HEIGHT:
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

# setting up sprites        
P1 = Player()
E1 = Enemy()

# creating sprite groups
enemies = pygame.sprite.Group()
enemies.add(E1)

all_sprites = pygame.sprite.Group()
all_sprites.add(P1)
all_sprites.add(E1)

coins = pygame.sprite.Group()
for _ in range(2):
    coin = Coin()
    coins.add(coin)

# adding a new User event 
INC_SPEED = pygame.USEREVENT + 1
pygame.time.set_timer(INC_SPEED, 1000)

# game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == INC_SPEED:
            SPEED += 0.5     
        if event.type == QUIT:
            running = False

    DISPLAYSURF.blit(background, (0, 0))

    # displaying score
    scores = font_small.render(f"Coins: {SCORE}", True, BLACK)
    DISPLAYSURF.blit(scores, (10, 10))

    # move and draw all sprites
    for entity in all_sprites:
        DISPLAYSURF.blit(entity.image, entity.rect)
        entity.move()

    for coin in coins:
        DISPLAYSURF.blit(coin.image, coin.rect)
        coin.move()

    # check for coin collection
    collected_coins = pygame.sprite.spritecollide(P1, coins, True)
    SCORE += len(collected_coins)

    # ensure at least 3 coins remain on screen
    while len(coins) < 3:
        coins.add(Coin())

    # check for collision with enemy
    if pygame.sprite.spritecollideany(P1, enemies):
        if os.path.exists("crash.wav"):
            pygame.mixer.Sound('crash.wav').play()
        time.sleep(0.5)

        DISPLAYSURF.fill(RED)
        DISPLAYSURF.blit(game_over, (30, 250))
        pygame.display.update()
        
        time.sleep(2)
        running = False  # Correct way to exit game loop

    pygame.display.update()
    FramePerSec.tick(FPS)

pygame.quit()
print("Game exited successfully.")

```

```python
import pygame, sys, time, random

# difficulty
difficulty = 10

level = 1  # starting level
score_to_next_level = 4  # points needed to level up

# window size
frame_size_x = 720
frame_size_y = 480

# initialize pygame
pygame.init()
pygame.display.set_caption('Snake')
game_window = pygame.display.set_mode((frame_size_x, frame_size_y))

# define colors
black = pygame.Color(0, 0, 0)
white = pygame.Color(255, 255, 255)
red = pygame.Color(255, 0, 0)
green = pygame.Color(0, 255, 0)

# fps controller
fps_controller = pygame.time.Clock()

# snake initial position and body
snake_pos = [100, 50]
snake_body = [[100, 50], [90, 50], [80, 50]]

# function to generate food position (not on the snake)
def spawn_food():
    while True:
        food_x = random.randint(0, (frame_size_x // 10) - 1) * 10
        food_y = random.randint(0, (frame_size_y // 10) - 1) * 10
        if [food_x, food_y] not in snake_body:
            return [food_x, food_y]

food_pos = spawn_food()

direction = 'RIGHT'
change_to = direction
score = 0

# function to handle game over
def game_over():
    global running
    my_font = pygame.font.SysFont('times new roman', 90)
    game_over_surface = my_font.render('Game Over', True, red)
    game_over_rect = game_over_surface.get_rect(midtop=(frame_size_x/2, frame_size_y/4))
    game_window.fill(black)
    game_window.blit(game_over_surface, game_over_rect)
    show_score_level()
    pygame.display.flip()
    time.sleep(3)
    running = False  # Exit the game loop

# display score and level
def show_score_level():
    score_font = pygame.font.SysFont('consolas', 20)
    score_surface = score_font.render(f'Score: {score}  Level: {level}', True, white)
    score_rect = score_surface.get_rect(midtop=(frame_size_x/2, 15))
    game_window.blit(score_surface, score_rect)

# main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and direction != 'DOWN':
                change_to = 'UP'
            if event.key == pygame.K_DOWN and direction != 'UP':
                change_to = 'DOWN'
            if event.key == pygame.K_LEFT and direction != 'RIGHT':
                change_to = 'LEFT'
            if event.key == pygame.K_RIGHT and direction != 'LEFT':
                change_to = 'RIGHT'
            if event.key == pygame.K_ESCAPE:
                running = False  # Allow exit with Escape

    # update direction
    direction = change_to

    # move the snake
    if direction == 'UP':
        snake_pos[1] -= 10
    elif direction == 'DOWN':
        snake_pos[1] += 10
    elif direction == 'LEFT':
        snake_pos[0] -= 10
    elif direction == 'RIGHT':
        snake_pos[0] += 10

    # snake growth logic
    snake_body.insert(0, list(snake_pos))
    if snake_pos == food_pos:
        score += 1
        food_pos = spawn_food()
        
        if score % score_to_next_level == 0:
            level += 1
            difficulty += 5
    else:
        snake_body.pop()

    # check for collisions with walls
    if snake_pos[0] < 0 or snake_pos[0] >= frame_size_x or snake_pos[1] < 0 or snake_pos[1] >= frame_size_y:
        game_over()
    
    # check for collisions with itself
    if snake_pos in snake_body[1:]:
        game_over()

    # draw game elements
    game_window.fill(black)
    for pos in snake_body:
        pygame.draw.rect(game_window, green, pygame.Rect(pos[0], pos[1], 10, 10))
    pygame.draw.rect(game_window, white, pygame.Rect(food_pos[0], food_pos[1], 10, 10))

    # display score and level
    show_score_level()

    # update display
    pygame.display.update()

    # set game speed
    fps_controller.tick(difficulty)

pygame.quit()
print("Game exited successfully.")
```

```python
#lab task3
import pygame

# initialize pygame
pygame.init()

# set up display
WIDTH, HEIGHT = 640, 480
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Drawing Tool")

# colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# variables
drawing = False
last_pos = None
color = BLUE
radius = 5
mode = "free_draw"  # free_draw, rectangle, circle, eraser
shapes = []  # stores drawn shapes

clock = pygame.time.Clock()

running = True
while running:
    screen.fill(BLACK)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        # handle keypresses for mode selection
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_r:
                color = RED
            elif event.key == pygame.K_g:
                color = GREEN
            elif event.key == pygame.K_b:
                color = BLUE
            elif event.key == pygame.K_e:
                mode = "eraser"
            elif event.key == pygame.K_f:
                mode = "free_draw"
            elif event.key == pygame.K_t:
                mode = "rectangle"
            elif event.key == pygame.K_c:
                mode = "circle"
            elif event.key == pygame.K_x:  # clear screen
                shapes.clear()
                screen.fill(BLACK)
        
        # handle mouse events
        if event.type == pygame.MOUSEBUTTONDOWN:
            drawing = True
            last_pos = event.pos
            if mode in ["rectangle", "circle"]:
                shapes.append((mode, color, last_pos, last_pos))
        
        if event.type == pygame.MOUSEBUTTONUP:
            drawing = False
            last_pos = None
        
        if event.type == pygame.MOUSEMOTION and drawing:
            if mode == "free_draw":
                pygame.draw.circle(screen, color, event.pos, radius)
                shapes.append(("dot", color, event.pos, radius))
            elif mode == "eraser":
                pygame.draw.circle(screen, BLACK, event.pos, radius)
                shapes.append(("dot", BLACK, event.pos, radius))
            elif mode in ["rectangle", "circle"]:
                shapes[-1] = (mode, color, shapes[-1][2], event.pos)
    
    # draw stored shapes
    for shape in shapes:
        shape_mode, shape_color, start_pos, end_pos = shape
        if shape_mode == "rectangle":
            pygame.draw.rect(screen, shape_color, (*start_pos, end_pos[0] - start_pos[0], end_pos[1] - start_pos[1]), 2)
        elif shape_mode == "circle":
            center = ((start_pos[0] + end_pos[0]) // 2, (start_pos[1] + end_pos[1]) // 2)
            radius = max(abs(end_pos[0] - start_pos[0]), abs(end_pos[1] - start_pos[1])) // 2
            pygame.draw.circle(screen, shape_color, center, radius, 2)
        elif shape_mode == "dot":
            pygame.draw.circle(screen, shape_color, 
                               start_pos, end_pos)
    
    pygame.display.flip()
    clock.tick(60)


pygame.quit()
```
